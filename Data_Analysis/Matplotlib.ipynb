### 01. Matplotlib 데이터 시각화

1. Matplotlib 기초
    * Matplotlib은 막대 그래프, 산점도 등 데이터 시각화 기능을 가진 대표적인 파이썬 라이브러리

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [1, 2, 3, 4, 5]
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title("My First Plot")
ax.set_xlabel("x")
ax.set_ylabel("y")
fig.set_dpi(70)
# fig.savefig("first_plot.png")
```

2. Line Graph
    * marker : 그래프의 점 모양
    * color : 그래프의 색상
    * linestyle : 그래프의 선 모양
    
| linestyle 문자  | 의미  |
|:-:|---|
|'-' | 	실선  |
| '-’ |  선이 띄엄띄엄 연결된 모양 |
| '-.'  | 선과 점이 연결된 모양  |
| ':'  | 점선  |

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(10)
fig, ax = plt.subplots()
ax.plot(
    x, x, label='y=x',
    marker='o',
    color='blue',
    linestyle=':'
)
ax.plot(
    x, x**2, label='y=x^2',
    marker='^',
    color='red',
    linestyle='--'
)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.legend(
    loc='upper left',
    shadow=True,
    fancybox=True,
    borderpad=2
)

fig.savefig("plot.png")
```

3. Scatter (산점도)
    * 이를 사용하게 되면 값 사이의 관계를 파악하기 용이해지고 데이터 셋에서 이상값을 찾기도 용이하다.
    * 둘 이상의 측정값의 상관 관계를 파악하고 시각화
        * markersize : 점의 크기
        * markerfacecolor : 점의 채우기 색상
        * markeredgecolor : 점의 테두리 색상
    > ax.plot(x축, y축, 마커, markersize, markerfacecolor, markeredgecolor)

```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
x = np.arange(10)
ax.plot(
    x, x**2, "o",
    markersize=15,
    markerfacecolor='white',
    markeredgecolor="blue"
)

fig.savefig("plot.png")
```

4. Bar & Histogram
    * Bar그래프(막대형 차트)
        * 여러 값을 비교하는데 적합, 여러개의 값을 입력 받고 그 값들을 한눈에 비교 할 수 있다.
    * Histogram
        * 일정 시간 동안의 숫자 데이터 분포를 시각화 하는데 적합

```python
# Bar 그래프
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(10)
fig, ax = plt.subplots(figsize=(8, 4))
ax.bar(x, x*2)

fig.savefig("plot.png")
```
