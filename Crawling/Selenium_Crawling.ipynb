### 01. 웹 스크레핑과 Selenium

### 02. 태그 이름으로 요소 찾기
* 코드 에러 발생 시 with 표현식을 사용하면 꺼지지 않은 브라우저가 메모리를 점유하는 현상을 방지할 수 있다.


* 브라우저로부터 태그명을 이용하여 요소를 찾는 method : find_element_by_tag_name('tag')
    * 복수 선택 : find_elements_by_tag_name('tag')

````python
from selenium import webdriver

cpath = './chromedriver.exe'

with webdriver.Chrome(executable_path=cpath) as driver:    # 실행 브라우저
    driver.get('https://news.naver.com')
    title = driver.find_element_by_tag_name('title')
    print(title.text)
````

### 04. 요소의 주소, XPath
 
1. DOM (Document Object Model : 문서 내의 모든 요소를 각 노드 간의 계층 구조로 표현함.


2. XPath : XML 문서의 특정 노드에 접근하기 위한 질의어


 * 문법 
    * / : (절대경로) 현재 위치의 자식 노드만 검색 
    * // : (상대경로) 현재 위치의 모든 자손 노드에서 검색 ex) //div/p
    * \* : 와일드 카드, 경로에 있는 모든 노드를 의미
    * [] : 필터 표현식, 인덱스와 속성 등을 통해 특정 요소를 검색할 수 있다.
      * [index] : 여타 언어와 달리 1부터 시작  ex) //p[1]
      * [@attr] : 검색한 노드들 중 해당 속성을 가지고 있는 노드를 모두 반환 

 
 * 활용
    * ex) //p[@class="bold"] : 모든 p 중에서 class 속성을 가지고 있는 데 속성 값이 bold인 것만 가지고 온다.
    * ex) //*[@class='big']  : 모든 노드에서 class 속성을 가지고 있는 데 속성 값이 big인 것만 가지고 온다.
    
 * Tip
   * xpath 에서는 class를 지정해주면 그 값만 가진 노드를 찾아오기 때문에 contains()를 사용
   * '//li[contains(@class, "big")]' : li 중, class 속성에 'big'이 포함되어 있는 경우

##  < Selenium 심화 : 브라우저 제어 >

### 01. Selenium을 이용한 브라우저 제어


1. 로그인 자동화 : 로그인 후에만 나오는 웹 페이지 분석을 위해 ID와 비밀번호 입력 후 로그인 버튼 클릭
2. 드랍다운 버튼 클릭 : 드랍다운 버튼을 클릭해야만 나오는 요소의 추출을 위해 드랍다운 버튼을 찾아서 클릭

### 02. 브라우저 로딩 기다리기


1. 무조건 기다리기 : **time.sleep(n)** , n초만큼 무조건 기다림 , 비효율적
2. 암시적 기다리기 : **implicitly_wait(n)** , 해당 요소의 로딩이 끝나면 즉시 기다리기를 종료하고 코드 수행, 한번 설정 -> 해당 브라우저 계속 적용
3. 명시적 기다리기 : 명시한 요소가 불러와 질때까지 최대 n초 기다림

```python
e = webDriverWait(driver,30).until(
    EC.presence_of_element_located((By.ID, "id_name"))
)
```

### 03. 키보드/마우스 입력

### 04. 다양한 입력, ActionChains

* ActionChains 사용법

    1) 
    ```python
    Chains = webdriver.ActionChains(driver)
    Chains.action1()
    Chains.action2()
    ...
    Chains.perform()
    ```
    
    2)
    ```python
    webdriver.ActionChains(driver)
    .action1()
    .action2()
    ...
    .perform()
    ```

<br>

1) send_keys_to_element
 * 주어진 요소에 주어진 key sequence를 입력
 * 하나의 ActionChains 내에서 여러 요소에 액션을 수행할 수 있음

2) send_keys
 * 요소가 주어지지 않기 때문에, 현재 커서가 위치한 곳에서 key sequence를 입력
    
3) move_to_element
 * 주어진 요소로 마우스 커서를 이동
    
4) pause
 * n초만큼 멈췄다 뒤의 액션을 수행
    
5) move_by_offset
 * x, y축으로 주어진 수치만큼 이동
 
 
* 예시
```python
# 요소 찾기
_id = driver.find_element_by_xpath('//*[@id="id"]')
_pw = driver.find_element_by_xpath('//*[@id="pw"]')
```
```
# 액션 수행
webdriver.ActionChains(driver)
    .send_keys_to_element(_id,'my_id')
    .send_keys_to_element(_pw,'my_pw')
    .send_keys(Keys.ENTER)
    .perform()
```
